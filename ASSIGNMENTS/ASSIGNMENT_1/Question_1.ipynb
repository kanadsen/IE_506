7.  Using T1 as training data, solve an ordinary least squares regression problem and obtain the regression coefficients. Using these coefficients, compute the R2 and mean squared error values on the training set T1 and test set T2

```python
response_cols=['Weight']
predictor_columns=dataset.drop(columns=['Weight'])  # Extracting the predictor columns
#print(predictor_columns,len(predictor_columns))
T1_train=T1_set.drop(columns=response_cols,axis=1) 
T2_test=T2_set.drop(columns=response_cols,axis=1) 
T1_train
```

```python
# converting predictor variables to numpy array
T1_train=T1_train.to_numpy()
T2_test=T2_test.to_numpy()
print(T1_train.shape,T2_test.shape)
```

```python
# Add a column of 1's to account for the beta0 in the matrix multiplication Y=beta0 + beta1*x1+....
X_train=np.hstack((T1_train, np.ones((T1_train.shape[0], 1), dtype=T1_train.dtype)))
X_test=np.hstack((T2_test, np.ones((T2_test.shape[0], 1), dtype=T2_test.dtype)))
print(X_train.shape,X_test.shape)
```

```python
# Convert respective response variables into numpy arrays
Y_train=T1_set[response_cols].to_numpy()
Y_test=T2_set[response_cols].to_numpy()
print(Y_train.shape,Y_test.shape)
```

```python
#Finding beta values

XTX = np.matmul(np.transpose(X_train),X_train)
#print('XTX shape:',XTX.shape)
#condition number of XTX
print('Condition Number',np.linalg.cond(XTX))

Xy = np.matmul(np.transpose(X_train),Y_train)
#print('Xy shape:',Xy.shape)

beta = np.matmul(np.linalg.inv(XTX),Xy)

print('Beta values',beta.shape)
print(beta)
```

```python
# R**2 value and MSE value of Train set

#Compute y_bar 
y_bar = np.mean(Y_train)
#print('y_bar of Train set',y_bar)

# Computing variance of y
sigma_yy = 0
for i in range(Y_train.shape[0]): #access each row from the data set 
  y_i = Y_train[i] 
  sigma_yy += (y_i-y_bar)**2

#print('sigma_yy of of Train set:', sigma_yy)

#Compute sum of squared residuals 
sum_sq_residuals = 0
for i in range(Y_train.shape[0]): #access each row from the data set 
  x_i = X_train[i,:] #access i-th row of X
  y_i = Y_train[i] #access i-th row of y

  y_pred_i = np.dot(x_i,beta) #Computing the prediction obtained using the regression coefficients 
  e_i = y_i - y_pred_i #Difference between the actual observation y_i and prediction y_pred_i

  sum_sq_residuals += (e_i)**2

print('Sum of squared residuals:', sum_sq_residuals)

#Then we will compute the R^2 quantity
R_sq = 1-sum_sq_residuals/sigma_yy
print('R**2 value for Train set:',R_sq)
# Mean Squared error
print("Mean Squared Error (MSE) of Train set =",sum_sq_residuals/X_train.shape[0])
```

```python
# R**2 value of Test set

#Compute y_bar 
y_bar = np.mean(Y_test)
#print('y_bar of Test set',y_bar)

# Computing variance of y
sigma_yy = 0
for i in range(Y_test.shape[0]): #access each row from the data set 
  y_i = Y_test[i] 
  sigma_yy += (y_i-y_bar)**2

#print('sigma_yy of Test set:', sigma_yy)

#Compute sum of squared residuals 
sum_sq_residuals = 0
for i in range(Y_test.shape[0]): #access each row from the data set 
  x_i = X_test[i,:] #access i-th row of X
  y_i = Y_test[i] #access i-th row of y

  y_pred_i = np.dot(x_i,beta) #Computing the prediction obtained using the regression coefficients 
  e_i = y_i - y_pred_i #Difference between the actual observation y_i and prediction y_pred_i

  sum_sq_residuals += (e_i)**2

print('Sum of squared residuals:', sum_sq_residuals)

#Then we will compute the R^2 quantity
R_sq = 1-sum_sq_residuals/sigma_yy
print('R**2 value for Train set:',R_sq)
# MSE
print("Mean Squared Error (MSE) of Train set =",sum_sq_residuals/X_test.shape[0])
```

Now, to perform â„“2 regularized ordinary least squares regression, choose regularization constant Î» âˆˆ {10âˆ’5, 10âˆ’4, 10âˆ’3, 10âˆ’2, 10âˆ’1, 1, 10, 100, 1000, 10000} using a 5-fold crossvalidation on the full train set T1. Use average R2 scores of the folds as the criterion to choose best regularization parameter. Perform â„“2 regularized ordinary least squares regression with best regularization parameter obtained from the  cross-validation procedure on the set T1 and obtain the regression coefficients. Using these coefficients, compute the R2 and mean squared
error values on the training set T1 and test set T2.

```python
# Function to compute R2 and MSE

def compute_Rsquared_and_MSE(beta, X, Y): 

  #first we shall compute y_bar 
  y_bar = Y.mean()
  #print('y_bar',y_bar)

  sigma_yy = 0
  for i in range(Y.shape[0]): 
       y_i = Y[i] 
       sigma_yy += (y_i-y_bar)**2
  #print('sigma_yy:', sigma_yy)

  #print('\n_________________________\n')

  #now we can compute sum of squared residuals 
  sum_sq_residuals = 0.
  for i in range(Y.shape[0]): #access each row from the data set 
    x_i = X[i,:] #access i-th row of X
    y_i = Y[i] #access i-th row of y

    y_pred_i = np.dot(x_i,beta) #compute the prediction obtained using the regression coefficients 
    e_i = y_i - y_pred_i #compute the difference between the actual observation y_i and prediction y_pred_i

    sum_sq_residuals += (e_i)**2

    
  #Then we will compute the R^2 quantity and MSE
  R_sq = 1-sum_sq_residuals/sigma_yy
  MSE = sum_sq_residuals/Y.shape[0]
  #print('R^2 :',R_sq, ' MSE:',MSE)
  return R_sq, MSE
```

```python
import scipy
from scipy import linalg

#print(T1_set.shape)

seeds = [100, 200,300,400,500] # Since 5 fold cross validation is chosen so 5 seeds are given

lambda_val = [pow(10,-5),pow(10,-4),pow(10,-3),pow(10,-2),pow(10,-1),1,10,100,1000,10000]

train_R2_seeds = [] 
val_R2_seeds = [] 

train_MSE_seeds = [] 
val_MSE_seeds = [] 
for seed in seeds: 
  T1_train = T1_set.sample(frac=0.8,random_state=seed) # since 5 five fold cross validation so dataset is divided into 5 chunks. So validation frac=0.2
  T1_val = T1_set.drop(T1_train.index)
  #print(T1_train.shape,T1_val.shape)
  #print(T1_train.Weight)
  X_val = T1_val.drop(columns=response_cols,axis=1)
  X_train = T1_train.drop(columns=response_cols,axis=1)
  

  X_train = X_train.to_numpy()
  X_val = X_val.to_numpy()

  X_train = np.hstack((X_train, np.ones((X_train.shape[0], 1), dtype=X_train.dtype)))
  X_val = np.hstack((X_val, np.ones((X_val.shape[0], 1), dtype=X_val.dtype)))
  
  Y_train = T1_train[response_cols].to_numpy()
  Y_val = T1_val[response_cols].to_numpy()

  train_R2_lambdas = []
  val_R2_lambdas = []

  train_MSE_lambdas = []
  val_MSE_lambdas = []
  
  for lambda_ in lambda_val:
    XTX = np.matmul(np.transpose(X_train),X_train)
    Xy = np.matmul(np.transpose(X_train),Y_train)
    I = np.identity(X_train.shape[1])

    beta =scipy.linalg.solve( XTX+lambda_*I,Xy)
    #print('lambda:',lambda_)
    #print('beta:',beta)
    #print('***************')

    train_R2, train_MSE = compute_Rsquared_and_MSE(beta, X_train, Y_train)
    train_R2_lambdas.append(train_R2)
    train_MSE_lambdas.append(train_MSE)

    val_R2, val_MSE = compute_Rsquared_and_MSE(beta, X_val, Y_val)
    val_R2_lambdas.append(val_R2)
    val_MSE_lambdas.append(val_MSE)

  train_R2_seeds.append(np.array(train_R2_lambdas))
  train_MSE_seeds.append(np.array(train_MSE_lambdas))

  val_R2_seeds.append(np.array(val_R2_lambdas))
  val_MSE_seeds.append(np.array(val_MSE_lambdas))

val_R2_seeds = np.array(val_R2_seeds).squeeze()
print(val_R2_seeds)

mean_val_R2_lambdas = np.mean(val_R2_seeds,axis=0)

train_R2_seeds = np.array(train_R2_seeds).squeeze()
print(train_R2_seeds)

mean_train_R2_lambdas = np.mean(train_R2_seeds,axis=0)

print('5-fold train R2 values:', train_R2_seeds,'\n')

print('5-fold train MSEs:', train_MSE_seeds,'\n')

print('5-fold val R2 values:', mean_val_R2_lambdas.squeeze(),'\n')
print('5-fold train R2 values:', mean_train_R2_lambdas.squeeze(),'\n')

#print('k-fold val MSEs:', val_MSE_seeds)


best_lambda_idx = np.argmax(mean_val_R2_lambdas)
best_lambda = lambda_val[best_lambda_idx]

print('best lambda:', best_lambda)

```

```python
# Obtaining beta fron the best lamda

X_train = T1_set.drop(columns=response_cols)
X_train = X_train.to_numpy()

X_train = np.hstack((X_train, np.ones((X_train.shape[0], 1), dtype=X_train.dtype)))
Y_train = T1_set[response_cols].to_numpy()

XTX = np.matmul(np.transpose(X_train),X_train)
Xy = np.matmul(np.transpose(X_train),Y_train)
I = np.identity(X_train.shape[1])

final_beta =scipy.linalg.solve( XTX+best_lambda*I,Xy)
print('lambda:',best_lambda)
print('Final beta:',final_beta.shape,'\n',final_beta)
#print('***************')
```

```python
X_train=T1_set.drop(columns=response_cols).to_numpy()
X_train = np.hstack((X_train, np.ones((X_train.shape[0], 1), dtype=X_train.dtype)))
Y_train=T1_set[response_cols].to_numpy()
X_test=T2_set.drop(columns=response_cols).to_numpy()
X_test = np.hstack((X_test, np.ones((X_test.shape[0], 1), dtype=X_train.dtype)))
Y_test=T2_set[response_cols].to_numpy()

train_R2, train_MSE = compute_Rsquared_and_MSE(final_beta, X_train, Y_train)
test_R2, test_MSE = compute_Rsquared_and_MSE(final_beta, X_test, Y_test)

print('Train R^2 :', train_R2, ' Train MSE:', train_MSE)
print('Test R^2 :', test_R2, ' Test MSE:', test_MSE)
```
