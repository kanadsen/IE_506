7.  Using T1 as training data, solve an ordinary least squares regression problem and obtain the regression coefficients. Using these coefficients, compute the R2 and mean squared error values on the training set T1 and test set T2

```python
response_cols=['Weight']
predictor_columns=dataset.drop(columns=['Weight'])  # Extracting the predictor columns
#print(predictor_columns,len(predictor_columns))
T1_train=T1_set.drop(columns=response_cols,axis=1) 
T2_test=T2_set.drop(columns=response_cols,axis=1)
T1_train
```

```python
T2_test
```

```python
# converting predictor variables to numpy array
X_train=T1_train.to_numpy()
X_test=T2_test.to_numpy()
Y_train=T1_set[response_cols].to_numpy()
Y_test=T2_set[response_cols].to_numpy()
print(X_train.shape,X_test.shape)
```

To perform â„“1 regularized least squares regression, choose regularization constant Î» âˆˆ {10âˆ’5, 10âˆ’4, 10âˆ’3, 10âˆ’2, 10âˆ’1, 1, 10, 100, 1000, 10000} using a 5-fold cross-validation on the full train set T1. Use average R2 scores of the folds as the criterion to choose best regularization parameter Î» â‹†
. Perform â„“1 regularized least squares regression with best regularization parameter Î» â‹† obtained from the cross-validation procedure on the set T1 and obtain the regression coefficients.

```python
# Performing L1 regression or lasso regression

# Here I am using grid search CV to search through the parameters in this case
from sklearn.model_selection import validation_curve
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
lambda_val = [pow(10,-5),pow(10,-4),pow(10,-3),pow(10,-2),pow(10,-1),1,10,100,1000,10000]
lasso=Lasso()
parameters={'alpha':lambda_val}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='r2',cv=5)  
# cv=5 because 5 fold cross validation is given
# scoring 'r2' is for averaged r2 score of the parameter
train_scores, valid_scores = validation_curve(lasso,X_train,Y_train, param_name="alpha", param_range=lambda_val,cv=5)

lasso_regressor.fit(X_train,Y_train)
print("Best Parameter from Grid Search CV :",lasso_regressor.best_params_)
print("Validation avg R2 score :",lasso_regressor.best_score_)

best_param=lasso_regressor.best_params_['alpha'] 
```

```python
#print(train_scores.shape)
avg_train_scores = np.mean(train_scores,axis=1)
avg_val_scores = np.mean(valid_scores,axis=1)
print('average train scores :',avg_train_scores)
print('average val scores :',avg_val_scores)
```

```python
plt.figure(figsize =(12,7))

plt.plot(lambda_val,avg_train_scores,label = "Training Accuracy value")
plt.plot(lambda_val, avg_val_scores,label = "Validation Accuracy value")
plt.xlabel('alpha')
plt.ylabel('Avg. accuracy score ')
plt.title("Avg. accuracy score vs alphas")
plt.xscale('log')
plt.grid()
plt.legend()
plt.xticks(lambda_val)
plt.show()
```

```python
best_alpha = lambda_val[np.argmax(avg_val_scores)]
print('best alpha from 5 fold CV:',best_alpha)
```

### Since we are getting alpha value = 0.001 as the best parameter from both the results we take lamda value = 0.001  

```python
from sklearn import linear_model
clf=linear_model.Lasso(alpha=best_param)
clf.fit(X_train,Y_train)
beta_val=clf.coef_
print('Lasso regression coefficients: \n', beta_val)
```

```python
# Storing attribute and beta values in dictionary
original_dct={}
dct={}
c=0
for i in columns:
    if(i=='Weight'):
       break
    original_dct[i]=beta_val[c]
    dct[i]=abs(beta_val[c])
    c+=1
print(dct)
```
